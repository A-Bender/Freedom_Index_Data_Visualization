```python
combined_country_mappings = {}

# First, add all countries from region_mapping
for region, countries in region_mapping.items():
    for country in countries:
        combined_country_mappings[country] = region

# Now, add countries from additional_country_mappings, skip if they are already present
for country, region in additional_country_mappings.items():
    # Only add if the country is not already in the combined mappings
    if country not in combined_country_mappings:
        combined_country_mappings[country] = region

# Output combined mappings to verify
print("Combined Country Mappings:")
for country, region in combined_country_mappings.items():
    print(f"{country}: {region}")
```

```python
gdp.loc[:, 'Region'] = gdp['Country Name'].map(combined_country_mappings)

# Check if there are any countries in the DataFrame that still have NaN in 'Region'
unknown_countries = gdp[gdp['Region'].isna()]

# Print any countries that were not matched in the mapping
if not unknown_countries.empty:
    print("Countries with unknown regions (not matched):")
    print(unknown_countries['Country Name'].unique())
else:
    print("All countries matched successfully!")

# Save the updated DataFrame to a new CSV file (or overwrite the existing one)
gdp.to_csv('updated_gdp_data.csv', index=False)
```

```python
# Calculate standard deviation and get the top fluctuation countries
gdp_std_dev = gdp[['Country Name', '2013', '2014', '2015', '2016', '2017', '2018', '2019']].copy()
gdp_std_dev.set_index("Country Name", inplace=True)
gdp_std_dev = gdp_std_dev.std(axis=1).reset_index(name='Std Dev')
top_fluctuation_countries = gdp_std_dev.nlargest(10, 'Std Dev')['Country Name']

# Filter original GDP DataFrame for these countries
top_gdp_data = gdp[gdp['Country Name'].isin(top_fluctuation_countries)]

# Melt the DataFrame into long format
top_gdp_melted = top_gdp_data.melt(id_vars=['Country Name'], var_name='Year', value_name='GDP')

# Ensure 'Year' is treated as a string
top_gdp_melted['Year'] = top_gdp_melted['Year'].astype(str)

# Ensure 'GDP' values are numeric, handling any non-numeric values
top_gdp_melted['GDP'] = pd.to_numeric(top_gdp_melted['GDP'], errors='coerce')

# Drop any rows with NaN values in 'GDP'
top_gdp_melted = top_gdp_melted.dropna(subset=['GDP'])

# Create the line plot
plt.figure(figsize=(14, 8))
sns.lineplot(data=top_gdp_melted, x='Year', y='GDP', hue='Country Name', marker='o')

plt.title('GDP Fluctuations of Top 10 Countries with Highest Standard Deviation')
plt.xlabel('Year')
plt.ylabel('GDP')
plt.legend(title='Country Name', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()  # Adjust layout to make room for the legend
plt.show()
```
