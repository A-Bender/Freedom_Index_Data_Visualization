```python
combined_country_mappings = {}

# First, add all countries from region_mapping
for region, countries in region_mapping.items():
    for country in countries:
        combined_country_mappings[country] = region

# Now, add countries from additional_country_mappings, skip if they are already present
for country, region in additional_country_mappings.items():
    # Only add if the country is not already in the combined mappings
    if country not in combined_country_mappings:
        combined_country_mappings[country] = region

# Output combined mappings to verify
print("Combined Country Mappings:")
for country, region in combined_country_mappings.items():
    print(f"{country}: {region}")
```

```python
gdp.loc[:, 'Region'] = gdp['Country Name'].map(combined_country_mappings)

# Check if there are any countries in the DataFrame that still have NaN in 'Region'
unknown_countries = gdp[gdp['Region'].isna()]

# Print any countries that were not matched in the mapping
if not unknown_countries.empty:
    print("Countries with unknown regions (not matched):")
    print(unknown_countries['Country Name'].unique())
else:
    print("All countries matched successfully!")

# Save the updated DataFrame to a new CSV file (or overwrite the existing one)
gdp.to_csv('updated_gdp_data.csv', index=False)
```
